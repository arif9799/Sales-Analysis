```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import rcParams
from cycler import cycler 


import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

rcParams['figure.figsize'] = 17,8
rcParams['lines.linewidth'] = 1
```

```python
data = pd.read_csv(path,skip_blank_lines=True, index_col = 0)
```

```python
spends = ['facebook_newsfeed_spend', 'youtube_brand_spend', 'search_spend', 'youtube_performance_spend', 'newspaper_spend', 'tv_spend']
impressions = ['facebook_newsfeed_impressions', 'youtube_brand_impressions', 'search_clicks', 'youtube_performance_impressions', 'newspaper_readership', 'tv_gross_rating_points']


data['Total_spent'] = data[spends].sum(axis = 1)
data['Total_impressions'] = data[impressions].sum(axis = 1)

sales_spends = data['sales'].corr(data['Total_spent'])
sales_impressions = data['sales'].corr(data['Total_impressions'])
sales_spends
sales_impressions

data['Log_Total_spent'] = np.log2(data['Total_spent'])
data['Log_Total_impressions'] = np.log2(data['Total_impressions'])
```

```python
data['Log_Total_spent'].plot()

plt.tick_params(axis='x', which='both', bottom=False, labelbottom=False)
plt.tick_params(axis='y', which='both', left=False, labelleft=False) 
sb.despine(right = True, top = True)
plt.plot(data['date'], data['Log_Total_spent'])
```
